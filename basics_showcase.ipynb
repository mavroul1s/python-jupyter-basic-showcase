## 2. ⚙️ Code Cells — Variables & Arithmetic

```python
# Basic variables and arithmetic
name = "Jupyter"
year = 2024
pi = 3.14159

print(f"Hello from {name}!")
print(f"Year  : {year}")
print(f"Pi    : {pi:.2f}")
print(f"Pi²   : {pi**2:.4f}")
```

---
## 3. 📋 Lists & Loops

```python
fruits = ["🍎 Apple", "🍌 Banana", "🍇 Grape", "🍊 Orange"]

print("Fruit list:")
for i, fruit in enumerate(fruits, start=1):
    print(f"  {i}. {fruit}")
```

---
## 4. 🔧 Functions

```python
def greet(name, loud=False):
    """Return a greeting string."""
    message = f"Hello, {name}!"
    return message.upper() if loud else message

print(greet("Alice"))
print(greet("Bob", loud=True))
```

## 5. 📊 A Simple Plot

Jupyter renders **matplotlib** figures inline — no extra setup needed.

```python
import matplotlib.pyplot as plt
import math

x = [i * 0.1 for i in range(0, 63)]
y_sin = [math.sin(v) for v in x]
y_cos = [math.cos(v) for v in x]

plt.figure(figsize=(8, 3))
plt.plot(x, y_sin, label="sin(x)", linewidth=2)
plt.plot(x, y_cos, label="cos(x)", linewidth=2, linestyle="--")
plt.title("Sine & Cosine")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.tight_layout()
plt.show()
```
